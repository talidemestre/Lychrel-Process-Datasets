# Largest Steps


```python
final_pal = 10000

n = 0
largest_n = 0
largest_pal = 0
largest_found_pal = 0

n_final = 1000

title_text = "Frequency Of Digits in the First " + str(n_final) + " Steps of the Lychrel Process" 
image_title = "images/196DigitDistributionForN" + str(n_final)

pal_array = []

# steps through partial lychrel process
for init_pal in range(0, final_pal):
    n=0
    pal = init_pal
    while n < n_final:
        reverse = int(str(pal)[::-1])
        if pal == reverse:
            print("Palindrome found for " + str(init_pal) + " at " + str(n) +" steps with a value of " + str(pal) )
            if n > largest_n:
                largest_n = n
                largest_pal = init_pal
                largest_found_pal = pal
            n = n_final + 1

        else:
            pal = pal + reverse
            n+=1
    
with open("196steps.csv", "w") as output:
    output.write(str(pal_array))
```

```python
print("Largest palindrome found at " + str(largest_pal) + " with " + str(largest_n) + " steps, arriving at " + str(largest_found_pal))
```
