# Universal Digits
We want to see if there are any digits that are in all partial iterations of the Lychrel process applied to 196.

```python
pal = 43550 
n = 0
n_final = 100000

start_val = pal + int(str(pal)[::-1])

# generator, break condition removed as we can save computation time

def lychrel_generator(start) :
    a = start
    b = a + int(str(a)[::-1])
    while True :
        yield a
        a , b = b , b + int(str(b)[::-1])

```

```python
digits = list( map( str, list(range(0,10))))
universal_occurence = [True, True, True, True, True, True, True, True, True, True]

lychrel_gen = lychrel_generator(start_val)

i=0
for char in lychrel_gen:
    occured = [False, False, False, False, False, False, False, False, False, False]
    if i > n_final :
        break
    else:
        #print(i)
        digit_list = map(int, list(str(char)))
        for digit in digit_list:
                #print("increasing", digit, "by one")
                occured[digit] = True
        j = 0
        while j < len(occured):
            if (occured[j]) == False:
                universal_occurence[j] = occured[j]
                print("No", j, "not found in", char, "at index", i, ".")
            j+=1        
        i += 1

print(universal_occurence)
print("Terminated at index", i, ".")
print("The condition that there are any universal digits in a Lychrel series is", any(universal_occurence),".")
```

Thus we can see there are no universal digits in this series when starting from 196. Interestingly however it appears that after index 192 all digits are always found in every value, at least up to index . While we have not established that there is a single digit responsible for all this madness, it is interesting to note that nearly all kin values of 196 appear to include every digit.
