# Reverse Process
Rather than reverse and add, this notebook demonstrates reverse and subtract.

```python
pal = 196
n = 0
n_final = 1000

# generator, break condition removed as we can save computation time
def lychrel_generator(start) :
    a = start
    b = abs(a - int(str(a)[::-1]))
    while True :
        yield a
        a , b = b , abs(b - int(str(b)[::-1]))

lychrel_gen = lychrel_generator(pal)


title_text = "Frequency Of Digits for " + str(n_final) + " Steps of the Lychrel Process on 196" 
image_title = "images/196DigitDistributionForN" + str(n_final)
```

```python
data = pd.read_csv('datasets/LychrelTo100000Integers.txt',  header=None)
data = data.values.tolist()[0][1:]
```

```python

valid = True

worked = 0
didnt_work = 0

for lychrel in data:
    lychrel_gen = lychrel_generator(lychrel)
    i=0
    for num in lychrel_gen:
        if (i > n_final or num == int(str(num)[::-1])):
            found_text = "Palindrome found for " + str(lychrel) + " at  " + str(num) + " after " + str(i) + " steps."
            print(found_text if num == int(str(num)[::-1]) else "limit exceeded for " + str(lychrel))
            valid = num == int(str(num)[::-1])
            if valid:
                worked+=1
            else:
                didnt_work+=1
            break
        else:
            #print(i, num)
            i+=1
print(valid)
print("The proportion of success is", str(worked/(worked+didnt_work) ))
```

For the sample size of 10000 integers the success rate of this method is 0.947, yet by increasing it to 100000 the success rate drops to 0.0.540531561461794. It is interesting to note that of those that do work, almost all of them seem to converge to a number of all 9s (99, 999, 9999, etc) or 0s bookended by 9s (909, 9009, 90009, etc). There also seem to be a handful of instances of numbers converging on 5445 or 54945.

It is also interesting to note that this method works for every single 3 digit lychrel.

Interestingly, it would appear that many of those that dont work fall into the same thread of continually alternating between 2178 and 6534, as by the reverse process each of these numbers subtracts into the other (8712 - 2178 = 6534, while 6534 - 4356 = 2178). Another thread I found that only occurs towards the upper end of the set (still interspersed with the previous example) is the similar alternation between 65934 and 21978.

```python
n_final = 10000

#store lychrel numbers
revlychrel_list = []

#store results
step_list = []
result_list = []

for n in range(n_final):
    lychrel_gen = lychrel_generator(n)
    i=0
    for num in lychrel_gen:
        if (i > n_final or num == int(str(num)[::-1])):
            found_text = "Palindrome found for " + str(n) + " at  " + str(num) + " after " + str(i) + " steps."
            print(found_text if num == int(str(num)[::-1]) else "limit exceeded for " + str(n))
            valid = num == int(str(num)[::-1])
            if valid:
                step_list.append(i)
                result_list.append(num)
            else:
                revlychrel_list.append(n)
            break
        else:
            #print(i, num)
            i+=1
```

```python
print(revlychrel_list)
```

This is the list of Reverse Lychrels (numbers who do not form a palindrome when applied to the reverse process) .


```python
forward_lychrel_data = pd.read_csv('datasets/LychrelTo100000Integers.txt',  header=None)
forward_lychrel_data = forward_lychrel_data.values.tolist()[0][1:]

unique_revlychrel = []
both = []

for rev_lychrel in revlychrel_list:
    valid = True
    for lychrel in forward_lychrel_data:
        if lychrel == rev_lychrel:
            valid = False
    if valid:
        unique_revlychrel.append(rev_lychrel)
    else:
        both.append(rev_lychrel)
```

```python
print(unique_revlychrel)
```

This is the list of Reverse Lychrels who are not Lychrels (will generate a palindrome regularly but not in reverse).

```python
print(both)
```

This is the list of numbers that act as Lychrels for both the regular and reverse process will not generate a palindrome in either case.

It is perhaps interesting to consider that this last list of numbers constitue what I am terming the 'true lychrels'. If we consider the reverse process to be the 'negative steps' of the lychrel process (i.e. 196 -691 is the negative 1st step in the process, which can perhaps even be refuted by pointing out that while 196 + 691 = 887 the opposite is not true as 887 - 788 ≠ 196), then 196 does indeed form a palindrome by the Lychrel process. However the numbers above never form a palindrome whether the process is applied to them 'regularly' or in reverse, thus they are 'true' Lychrel numbers as defined by not forming a palindrome by the Lychrel process.
